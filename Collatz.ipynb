```python
def main(ini, fim):
    cache = {}
    for i in range (ini, fim + 1):
        x = i
        cont = 0
        trajetoria = []
        while x != 1 and x not in cache:
            trajetoria.append(x)
            if x % 2 == 0:
                x = x // 2
            else:
                x = (3*x + 1)
            cont += 1
        if i in cache:
            cont += cache[x]
        for idx, num in enumerate(trajetoria):
            cache[num] = cont - idx
        print(f"Número {i}: {cont} operações até chegar em 1")
```
